```python
import pandas as pd
import numpy as np
```

```python
casa = pd.read_excel("house.xlsx")
casa.head()
```

```python
casa.info()
```

```python
casa = casa.dropna()
```

```python
casa = pd.get_dummies(casa)
casa.head()
```

```python
labels = np.array(casa["SalePrice"])
features = casa.drop("SalePrice", axis=1)
feature_list = list(casa.columns)
casa  = np.array(features)
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features,labels, test_size = 0.25, random_state=42)
```

```python
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor

```

```python
rf = RandomForestRegressor(n_estimators=1000, random_state=42)
rf.fit(train_features, train_labels);
```

```python
predictions_rf = rf.predict(test_features)

errors_rf = abs(predictions_rf - test_labels)

r_sq = rf.score(features, labels)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels, predictions_rf))
print("MSE:", metrics.mean_squared_error(test_labels, predictions_rf))

```

```python
ada = AdaBoostRegressor(n_estimators=1000, random_state=42)
ada.fit(train_features, train_labels);
```

```python
predictions_ada = ada.predict(test_features)

errors_ada = abs(predictions_ada - test_labels)

r_sq = rf.score(features, labels)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels, predictions_ada))
print("MSE:", metrics.mean_squared_error(test_labels, predictions_ada))
```

```python
gbr = GradientBoostingRegressor(n_estimators=1000, random_state=42)
gbr.fit(train_features, train_labels);

predictions_gbr = gbr.predict(test_features)

errors_gbr = abs(predictions_gbr - test_labels)

r_sq = rf.score(features, labels)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels, predictions_gbr))
print("MSE:", metrics.mean_squared_error(test_labels, predictions_gbr))
```
